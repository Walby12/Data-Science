```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data_base = pd.read_csv("cities_temperature.csv")

x = data_base["City ID"]
y = data_base["Average Temperature (°C)"]

proc_x = x.values.reshape(-1,1)
proc_y = y.values.reshape(-1,1)

modello = LinearRegression()
modello.fit(proc_x, proc_y)

city_id = 6537
previsione = modello.predict([[city_id]])
print(f"L' id {city_id} della città ha una temperatura di {previsione}°C")

punteggio = modello.score(proc_x, proc_y)
print(f"score: {punteggio}")
```

```python
sns.scatterplot(x=x, y=y)
plt.xlabel("City ID")
plt.ylabel("Average Temp")
plt.title("Distribuzione ID città vs previsione (°C)")
plt.show()
```
